```python
def checkValidProb(df, index):
    if df[index].sum() == 1.0:
        return True
    else:
        return False

def marginalise(df, indexes):
    """Calculate marginal distribution.

    Args:
        df (Dataframe): Dataframe of probabilities.
        indexes (List): Indexes of variables in marginal.

    Returns:
        Dataframe: Marginal Distribution.
    """
    marginalised = set(df.columns.to_list()) -set(indexes) - set("P")
    return joint.groupby(indexes, as_index=False).sum().drop(marginalised.pop(), axis=1)

```
